### 豆知識．2-gram エッジリスト

```python
word = ['今日', 'は', '晴れ', 'です', '。']
bigram = []

for i in range(len(word)-1):
    bigram.append([word[i], word[i+1]])

print(bigram)
```

### it-life-hack 記事の 2gram モデル作成

```python
import os
import re
from janome.tokenizer import Tokenizer

wakati = []

# 分かち書きオブジェクト生成
t = Tokenizer(wakati=True)
# ファイルの取得
files = os.listdir('./data/it-life-hack/')
    
for file in files:
    # ファイルオープンと読み込み
    f = open('./data/it-life-hack/' + file, 'r', encoding='utf-8')
    text = f.read()
        
    # 正規表現で不要な文字列を除去して表示
    reg_text = re.sub(r'[0-9a-zA-Z]+', '', text)
    reg_text = re.sub(r'[:;/+\.-]', '', reg_text)
    reg_text = re.sub(r'[\s\n]', '', reg_text)
        
    # 分かち書きして単語をリストへ格納
    wakati.append(t.tokenize(reg_text))
                
    # ファイルクローズ
    f.close()
```

```python
# リストのサイズ確認
print(len(wakati))
# 1つ目の記事の単語を確認
print(wakati[0])
```

```python
import itertools
from collections import Counter
import pandas as pd

# 単語の出現数をカウント
word_freq = Counter(itertools.chain(* wakati))

# 単語を出現数降順で並べ替え
dic = []
for word_uniq in word_freq.most_common():
    dic.append(word_uniq[0])

# 単語数を確認
print(len(dic))
# 出現数の多い単語を5つ確認
print(pd.DataFrame(dic).head())
```

```python
# 単語にIDを付与し辞書を作成
dic_inv = {}
for i, word_uniq in enumerate(dic, start=1):
    dic_inv.update({word_uniq: i})
    
# 辞書の長さを確認
print(len(dic_inv))
```

```python
# 練習問題4
# 単語をIDへ変換
wakati_id = [ [ dic_inv[word] for word in waka ] for waka in wakati ]

# リストのサイズ確認
print(len(wakati_id))
# 1つ目の記事の単語IDを確認
print(wakati_id[0])
```

```python
tmp = []
bigram = []

for i in range(len(wakati_id)):
    row = wakati_id[i]
    for j in range(len(row)-1):
        tmp.append([row[j], row[j+1]])
    bigram.extend(tmp)
    tmp = []

# 2-gramの表示
print(bigram)
```

```python
bigram_df = pd.DataFrame(bigram)
bigram_df = bigram_df.rename(columns={0:'node1', 1:'node2'})

# 各単語ペアに重み1を付与
bigram_df['weight'] = 1
# 重複する単語ペアの重みを足し合わせ
bigram_df = bigram_df.groupby(['node1', 'node2'], as_index=False).sum()

# 重み10を超え500未満の単語ペアを抽出
bigram_df = bigram_df[bigram_df['weight'] > 10]
bigram_df = bigram_df[bigram_df['weight'] < 500]

# 2-gramの確認
bigram_df.head()
```
